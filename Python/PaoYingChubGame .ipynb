# **Welcome to Pao Ying Chub Game**

```python
import random

#value
round : int = 0
win  : int = 0
lose : int = 0
draw : int = 0

list = ['Rock', 'Paper', 'Scissor']
bot = random.sample(list, k = 1)

#intro
print("Welcome to Pao Ching Chub game")
print("Game Rule:")
print("You will play against Bot A ")
print("Rock wins Scissors ")
print("Scissors wins Paper")
print("Paper wins Rock ")
print("------------------------------------------------ ")
name = input("Enter Your name:")





#loop

while (True):
      print("------------------------------------------------ ")
      action = int(input("Choose your number -> 1.Rock, 2.Paper, 3.Scissor, 4.Exit: "))
      
      if action == 1: #rock
          round += 1
          bot = random.sample(list, k = 1)
          print('You Play:Rock')
          print(f'Bot play:{bot}')
          if bot == ['Rock']:
                print("Draw")
                draw += 1
          elif bot == ['Paper']:
                print("Lose")
                lose += 1
          elif bot == ['Scissor']:
                print("Win")
                win += 1
      elif action == 2: #paper
          round += 1
          bot = random.sample(list, k = 1)
          print('You Play: Paper')
          print(f'Bot play:{bot}')
          if bot == ['Rock']:
                print("Win")
                win += 1
          elif bot == ['Paper']:
                print("Draw")
                draw += 1
          elif bot == ['Scissor']:
                print("Lose")
                lose += 1
      elif action == 3 : #scrissor
          round += 1
          bot = random.sample(list, k = 1)
          print('You Play:Scrissor')
          print(f'Bot play:{bot}')
          if bot == ['Rock']:
                print("Lose")
                lose += 1
          elif bot == ['Paper']:
                print("Win")
                win += 1
          elif bot == ['Scissor']:
                print("Draw")
                draw += 1
      else: 
#result
          print(f"Player: {name}")   
          print(f"Round: {round}")   
          print(f"You win: {win}")
          print(f"You lose: {lose}")
          print(f"You draw: {draw}")   
          break


        
    
    

```
